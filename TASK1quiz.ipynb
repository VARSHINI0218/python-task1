```python
import tkinter as tk
from tkinter import messagebox
import random

# Define questions and answers
questions = {
    "Science": {
        "easy": [
            {
                "question": "What is the chemical symbol for water?",
                "options": ["H2O", "CO2", "NaCl", "O2"],
                "answer": "H2O"
            },
            {
                "question": "What is the largest organ in the human body?",
                "options": ["Heart", "Liver", "Brain", "Skin"],
                "answer": "Skin"
            },
            {
                "question":"What is the nearest star to Earth?",
                "option":[" Sun","Proxima Centauri", "Alpha Centauri"," Sirius"],
                "answer":"sun"
               }
            ],

        "medium": [
            {
                "question": "What is the powerhouse of the cell?",
                "options": ["Nucleus", "Mitochondria", "Ribosome", "Lysosome"],
                "answer": "Mitochondria"
            },
            {
                "question": "Which planet is known as the Red Planet?",
                "options": ["Venus", "Mars", "Jupiter", "Saturn"],
                "answer": "Mars"
            }
        ],
        "hard": [
            {
                "question": "What is Avogadro's number?",
                "options": ["6.02 x 10^23", "3.14 x 10^4", "9.81 x 10^-3", "1.61 x 10^2"],
                "answer": "6.02 x 10^23"
            },
            {
                "question": "What is the process by which plants make their food?",
                "options": ["Photosynthesis", "Respiration", "Fermentation", "Transpiration"],
                "answer": "Photosynthesis"
            }
        ]
    }
}

class QuizApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Quiz Game")

        self.topic_label = tk.Label(root, text="Choose a topic:")
        self.topic_label.pack()

        self.topic_var = tk.StringVar(root)
        self.topic_var.set("Science")  # Default topic
        self.topic_menu = tk.OptionMenu(root, self.topic_var, *questions.keys())
        self.topic_menu.pack()

        self.difficulty_label = tk.Label(root, text="Choose difficulty:")
        self.difficulty_label.pack()

        self.difficulty_var = tk.StringVar(root)
        self.difficulty_var.set("easy")  # Default difficulty
        self.difficulty_menu = tk.OptionMenu(root, self.difficulty_var, "easy", "medium", "hard")
        self.difficulty_menu.pack()

        self.start_button = tk.Button(root, text="Start Quiz", command=self.start_quiz)
        self.start_button.pack()

    def start_quiz(self):
        topic = self.topic_var.get()
        difficulty = self.difficulty_var.get()
        questions_list = questions[topic][difficulty]
        random.shuffle(questions_list)

        score = 0
        total_questions = min(len(questions_list), 5)  # Limiting to 5 questions
        for question_data in questions_list[:total_questions]:
            answer = self.ask_question(question_data)
            if answer == question_data["answer"]:
                score += 1

        messagebox.showinfo("Quiz Complete", f"Quiz complete! You scored {score}/{total_questions}.")

    def ask_question(self, question_data):
        answer = tk.StringVar()
        answer.set(None)

        top = tk.Toplevel(self.root)
        top.title("Question")

        question_label = tk.Label(top, text=question_data["question"])
        question_label.pack()

        for idx, option in enumerate(question_data["options"], 1):
            tk.Radiobutton(top, text=option, variable=answer, value=option).pack()

        submit_button = tk.Button(top, text="Submit", command=top.destroy)
        submit_button.pack()

        top.wait_window()  # Wait for the user to close the question window
        return answer.get()

if __name__ == "__main__":
    root = tk.Tk()
    app = QuizApp(root)
    root.mainloop()

```
